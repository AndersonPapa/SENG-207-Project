```python
#Assigning the index values to 2 different variables in a slightly different order so they can be plotted against each other
plotting= ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality']

plotting1=['volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality','fixed acidity']
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(35, 35))
#Plotting the various subplots/axes in a figure and assigning it 4 rows and 3 columns because of our 12 different plots to be made

for i,z in zip(enumerate(plotting),enumerate(plotting1)):
    
#iterating using the zip method the 2 different lists made above into a list of tuples with the enumerate method which assigns numbers to each values as
#though they were an nd array in the first 'column' index[0] of the array and leaving the string values(names of columns) in the second 'column' index[1]
    
    x=i[1]
    y=z[1]
    
#taking the string values(name of columns) in variables; plotting and plotting1, for iteration
    
    ax=axes.flatten()
    
#Flattening the 2 dimensional nd array so the plots can be easily made
    
    ax[i[0]].scatter(x, y, data=dataset)
#A scatter diagram of x and y values in the data set

    ax[i[0]].set(xlabel=f'{x}', ylabel=f'{y}')
    ax[i[0]].set_title(f'Scatter Plot of {x} vs {y}')

    
# Creating a trendline using numpy and matplotlib

    
    trendline_param = np.polyfit(dataset[f'{x}'].fillna(dataset[f'{x}'].mean()), dataset[f'{y}'].fillna(dataset[f'{y}'].mean()),1)
    
#Using polyfit function to create a function involving the polynomial of least fit using each pair of dataset in for loop to account more for errors than using a line of best fit
#Each pair of columns plotted against each other had to be checked for Nan values and be corrected in order for the poly function to be run.
#The Nan values were replaced with the mean of each iterable column values
    
    function = np.poly1d(trendline_param)
    
#Generating a random range between the minimum and maximum of the data set in each iteration to ensure the trendline fits on the graph
    
    e = np.linspace(dataset[f'{x}'].min() ,dataset[f'{x}'].max() )
    
#Plotting the values of the range against that of the poly1d to create the trendline
    
    ax[i[0]].plot(e, function(e), 'g--')
```
